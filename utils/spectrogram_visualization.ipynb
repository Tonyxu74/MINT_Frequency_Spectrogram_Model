```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('../data/train/trainfiles/10.npy',allow_pickle=True)
```

```python
zero, one, two, three, four = 0,0,0,0,0

for key in data.item():
    f=data.item()[key]
    if f == 0:
        zero = zero+1
    if f==1: 
        one = one+1
    if f==2: 
        two=two+1
    if f==3: 
        three = three+1
    if f==4: 
        four=four+1
        
print(zero)
print(one)
print(two)
print(three)
print(four)
```

```python
i=0
for key in data.item():
    f=data.item()[key]
    i = i+1
    if i > 450:
        print(f)
```

```python
dataravel = data[0]
#print(dataravel.shape)

a = np.arange(0,data.shape[1],1)

x = np.array([])
y = np.array([])

b = np.arange(0,12)

for i in range(data.shape[1]):
    a = i*np.ones(data.shape[1])
    x = np.concatenate((x,a))
    y = np.concatenate((y,b))

print(data.shape)
#print(x)
#print(y)
dataravel = dataravel.flatten()
#print(dataravel)
```

```python
fig, ax = plt.subplots()
h = ax.hist2d(y, x, weights=dataravel)
plt.colorbar(h[3], ax=ax)
plt.xlabel('Arbitrary Time Units', fontsize=12)
plt.ylabel('Arbitrary Frequency Units', fontsize=12)
plt.title('Relative Frequency Contributions of One Channel',fontsize=12)

fig.savefig('input_sample.png')
```

```python
import csv
wavelet = np.array([])

t=8

with open('../data/raw/sample_data.txt', newline='') as csvfile:
    reader = csv.reader(csvfile)

    
    i=0
    for row in reader:
        i=i+1
        if t*250 <i< (t+1)*250:
            #print(row[1])
            wavelet= np.append(wavelet, float(row[1])+20000)
            
            
time = np.arange(wavelet.shape[0])
#print(time)
            
fig = plt.plot(time,wavelet)
plt.xlabel('Arbitrary Time Units', fontsize=12)
plt.ylabel('Arbitrary Frequency Units', fontsize=12)
plt.title('Relative Frequency Contributions of One Channel',fontsize=12)

#plt.savefig('wavelet.png')
            

            
```
